```python
mydata = pd.read_csv("employee_data.csv")
mydata[['First Name','Last Name']] = mydata.Name.apply(lambda x: pd.Series(str(x).split(" ")))
mydata = mydata.drop(columns = ["Name"])
mydata['DOB'] = pd.to_datetime(mydata['DOB'])
mydata["DOB"] = pd.to_datetime(mydata["DOB"]).dt.strftime('%m/%d/%Y')
mydata["SSN"] = mydata['SSN'].str.replace(r'^\d{3}-\d{2}', "***-**", regex=True)
mydata["State"] = mydata["State"].map(us_state_abbrev)
mydata = mydata[["Emp ID", "First Name", "Last Name", "DOB", "SSN", "State"]]

mydata.head()
```
